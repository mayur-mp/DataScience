```python
#we want  to predict the salary  it is depend upont the experience as the experiecne increases the
#so the Y is salry which we want to predict
salary.columns
```

```python
#step 2
y=salary['Salary']
x=salary[['Experience Years']]
#square brackets are used to give the shapes
#this use the liner alegebra matrix multiplicaton shapes rules
```

```python
x.shape
y.shape
```

```python
x.shape
```

(traing testing and splitting is done here in Step 4)

```python
#STEP 4
#split now train and test data for the  model
#use train_test_split function
#from skleanr library we will import the model slection module and  train test split funtion
#70% data will be trained and the 3o% data will be testesd
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529 )
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
#step5
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#step6  Model traning
y_pred=model.fit(x_train,y_train)
y_pred
```

```python
#step7 now predit the y that is slary using predict function
#our model will use the math fucntion to pedict the salary ie is y=c+mx+e predict function uses this formula it can also use another formula
#where the y is slary and c is the intercept m is the coef. and x is expereince and e is error
#here we are testing the data x
y_pred=model.predict(x_test)   # predection for the test smaple data x

x_test
```
