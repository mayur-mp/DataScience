```python
#step1
import pandas as pd
house=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv')
house.head()
```

```python
house.info()
```

```python
house.columns
```

```python
#step2
#now we define x and y   y is depenednnt variable y is (MEDV)  media house prices  X is independent variables (13) x variables
# and y is dependent which we will predict that is 1  ie total 14 columns

y=house['MEDV']
x=house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#step3  (data for traing testing and splitting is done here )
#tain test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
#step4 model selction  and model traning
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
#we get 13 slopes due to 13 x
```

```python
#steep5  Model predcition
y_pred=model.predict(x_test) # predection for the test smaple data x
x_test.head()
```

```python
y_pred[0:5]
```

```python
#step6 accuracy
#step8 Check accuracy of the model
#we will find the absoultue% error and the - it form the 100% remaining will be the accuracy of the model
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)  #in this fucnion it is taking the difference between(ytest - ypred)
```
